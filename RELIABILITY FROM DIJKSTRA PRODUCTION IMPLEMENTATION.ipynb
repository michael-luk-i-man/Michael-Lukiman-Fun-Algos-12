```python
import math as m

Gv = [[1,0], [2,99], [3,99], [4,99]] # A B C D with distances impossibly high, except for source 1/A = 0. 99 represents inf in this case. 
Ge = [ # u, v, reliability
     [1, 2, 0.5], # 0
     [1, 3, 0.4], # 1
     [2, 4, 0.3], # 2
     [3, 4, 0.5]  # 3
     ]

first = 0 
second = 1
third = 2


#Succinct conversion to reliability task while keeping Djikstra as a black box: 

for e in Ge:
    e[third] = m.log(1/(e[third])) # Convert all weights via reciprocal then total log. 
    
for v in Gv:
    v[second] = m.log(m.e/ (1 - float(v[second]/100) )) # Convert all distances to reliabilities via log, maintaining linear order. Log is base e, so reliability of starting node is 100%. 
    
    
#--------


print(Ge)

Image(filename='graph.png')
```

```python

# Michael Lukiman

import heapq, logging
from IPython.display import Image

vpi = ['nil']*len(Gv)
Gadj = [[] for n in range(len(Gv))]

def w(u,v):
    for e in Ge:
        if e[first] == u or e[second] == u: 
            if e[second] == v or e[first] == v:
                return e[third]
            
print('\n\nCreating the adjacency lists\n--------------------')
for e in Ge:
    print('\nEdge:', e)
    for vertex in Gv:
        print('\tVertex:',vertex)
        if e[first] == vertex[first]: 
            u = e[first]
            v = e[second]
            print('\t\tu:',u, ' | v:', v)
            Gadj[u-1].append(v)
            Gadj[v-1].append(u)
            print('\t\t',Gadj)
            
# print(vd, vp, Gadj)

for v in Gv:
    key = v[first]
    v[first] = v[second]
    v[second] = key


Q = Gv
heapq.heapify(Q)

distance = []

for vertex in Gv:
    distance.append(vertex[first])
    
S = [] 

def relax(u,v,distance):
    print('Adj:',v)
    print('Distance', v, ':',distance[v-1], '| Distance', u[1], ':',distance[u[1]-1], ' | w(',u[1],',',v,'):',w(u[1],v))
    if distance[v-1] > distance[u[1]-1] + w(u[1],v):
        distance[v-1] = distance[u[1]-1] + w(u[1],v)
        print(distance[v-1])
        vpi[v-1] = u


print('\n\nRelaxing the edges\n------------------')
while len(Q) != 0:
    u = heapq.heappop(Q)
    S.append(u)
    print(u)
    for v in Gadj[(u[1]-1)]:
        relax(u,v,distance)
        print('\n')

count = 0
for d in distance:
    count+=1
    print(count,':',d)

if distance[1-1]==0 and distance[2-1]==0.5 and distance[3-1]==0.4 and distance[4-1]==0.8:
    print('Got it!')

print('\n\nParent Map\n------------------')
print(vpi)
```
